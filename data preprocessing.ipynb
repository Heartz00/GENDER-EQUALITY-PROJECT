Transpose the data, to separate the data into three extra datasets consisting of the females, males and persons each.

```python
#transpose the data
df1 = df.T
df1.columns=df1.iloc[0]
df1.drop(['year'], axis=0, inplace=True)
df1
```

```python
#Extract the females features only from the datasets
df_females = df1[0:9]
df_females
```

```python
#save the female datasets in a csv format
df_females.to_csv('C:/Users/DELL/Desktop/Hamoyecapstoneproject/female earnings.csv')
```

```python
#Extract the male features only from the datasets
df_males = df1[9:18]
df_males
```

```python
#save the male datasets in a csv format
df_males.to_csv('C:/Users/DELL/Desktop/Hamoyecapstoneproject/male earnings.csv')
```

```python
#Extract the persons features only from the datasets
df_persons = df1[18:]
df_persons
```

```python
#save the persons datasets in a csv format
df_persons.to_csv('C:/Users/DELL/Desktop/Hamoyecapstoneproject/persons earnings.csv')
```
